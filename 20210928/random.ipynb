```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
numbers = np.random.logistic(0.0, 1.0, 50)
numbers
plt.hist(numbers, density=True, bins=6)
```

```python
numbers = np.random.normal(0.0, 1.0, 50)
numbers
plt.hist(numbers, density=True, bins=6)
```

```python
numbers = np.random.uniform(0.0, 1.0, 50)
numbers
plt.hist(numbers, density=True, bins=6)
```
